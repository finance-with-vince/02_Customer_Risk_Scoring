```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
customer_df = pd.read_csv('./customers.csv')
customer_df
```

```python
# Look at all the countres in the dataset
countries = customer_df['Country'].unique().tolist()
countries
```

```python
# Look at all ocupations held by customers
occupations = customer_df['Occupation'].unique().tolist()
occupations
```

```python
# Categorize countires and occupations
high_risk_countries = ['Cayman Islands', 'Panama', 'Mexico', 'USA']
high_risk_occupations = ['Casino Owner', 'Jewelry Dealer', 'Real Estate Agent']

medium_risk_countries = ['Switzerland', 'Canada', 'UK']
medium_risk_occupations = ['Accountant', 'Banker', 'Lawyer', 'Small Business Owner']

low_risk_occupations = ['Teacher', 'Doctor', 'Software Engineer']

# Define a function to rate countries
def country_score(row):
    country_score = 0
    if row['Country'] in high_risk_countries:
       country_score += 40
    elif row['Country'] in medium_risk_countries:
        country_score += 20

    return country_score

# Define a fuction to rate occupation
def occupation_score(row):
    occupation_score = 0
    if row['Occupation'] in high_risk_occupations:
        occupation_score += 40
    elif row['Occupation'] in medium_risk_occupations:
        occupation_score += 30
    elif row['Occupation'] in low_risk_occupations:
        occupation_score += 20

    return occupation_score

# Apply functions to new column
customer_df['Country_Risk_Score'] = customer_df.apply(country_score, axis=1)
customer_df['Occupation_Risk_Score'] = customer_df.apply(occupation_score, axis=1)

# Display
customer_df
```

```python
# # Determine function to define a the risk level based on occupation and country

def risk_level(row):
    total_score = row['Country_Risk_Score'] + row['Occupation_Risk_Score']

    if total_score >= 70:
        return 'High'
    elif total_score >=50:
        return 'Medium'
    else:
        return 'Low'
    
# Create Risk Level
customer_df['RiskLevel'] = customer_df.apply(risk_level, axis=1)

# Display
customer_df
```
